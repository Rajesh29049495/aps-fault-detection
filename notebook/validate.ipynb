```python
##below code is just to see hoe the tes works
import numpy as np
from scipy import stats
rng = np.random.default_rng()
sample1 = stats.uniform.rvs(size=100, random_state=rng)
sample2 = stats.norm.rvs(size=110, random_state=rng)
response = stats.ks_2samp(sample1, sample2)  ##this function will check whether these sample1 and smaple2 belongs to the same distribution, on the basis of it , it will calculate some values tha we have stored in the response
```

```python
sample1
```

```python
sample2
```

```python
response.pvalue   ##if my pvalue is less than 0.05, then we will reject null hypothesis{i.e., the distribution is different}
```

```python
round(response.pvalue,17) ##so basically it is very small, and ofcourse smaller than 0.05, therefore we reject the null hypothesis, i.e., these both samples are not from same distribution,,and it is very obvious as we have taken sample1 as uniform distribution and sample2 as normal distribution{this we have done just to understand how this function works}
```

**When we get pvalue like this, we say that the distribution of our data is changed, i.e., there is data drift, now we will have to do the retraining, i.e., the whole EDA,FE , then experiment with the models to get the most effective one need to be done.**
